## Obtenção dos Dados

```
# importar os pacotes necessarios
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns # bibliotecas para apresentação visual
import plotly.express  as  px
import plotly.io as pio
#plt.style.use('dark_background')

#configuração da biblioteca seaborn para os gráficos
sns.set(rc={'figure.figsize':(10, 6)}) # tamanho da área de plotagem
#sns.set_context("talk") # fontes maiores na visualização

# importar o arquivo listings.csv para um DataFrame
df = pd.read_csv("http://data.insideairbnb.com/south-africa/wc/cape-town/2021-07-18/visualisations/listings.csv")
```

# **Q6. Qual a correlação existente entre as colunas ?**
Correlação significa que existe uma relação entre duas coisas. No nosso contexto, estamos buscando relação ou semelhança entre as colunas.

Essa relação pode ser medida, e é função do coeficiente de correlação estabelecer qual a intensidade dela. Para identificar as correlações existentes entre algumas colunas, desta forma gerar:

* Uma matriz de correlação;

```
# criar uma matriz de correlação
corr = df_clean[['price', 'minimum_nights', 'number_of_reviews',
    'calculated_host_listings_count', 'availability_365']].corr()

display(corr)
```

# **Q7. Qual o tipo de imóvel mais alugado ?**
A coluna da variável room_type indica o tipo de locação que está anunciada no Airbnb. No site, existem opções de apartamentos/casas inteiras, apenas o aluguel de um quarto ou mesmo dividir o quarto com outras pessoas.

Vamos contar a quantidade de ocorrências de cada tipo de aluguel, usando o método value_counts().


```
# mostrar a quantidade de cada tipo de imóvel disponível
df_clean.room_type.value_counts()
```

```
##### FUNÇÃO CRIADA PARA CONSTRUIR O GRÁFICO DE BARRA ############ (NÃO ALTERAR)
def graficoBarra(colunaGrafico,tituloGrafico,rotuloEixoX,rotuloEixoY,legendaGrafico):
  # Montando o gráfico com o tipo de imóvel
  plt.figure(figsize = (8,5))
  # dados categóricos: Tipo de imóvel
  tipoImovel=df_clean[colunaGrafico].unique()
  # total por tipo de imóvel
  contTipoImovel=df_clean[colunaGrafico].value_counts()
  # parãmetros para construir o gráfico de barra
  plt.bar(tipoImovel, contTipoImovel, width= 0.8, align='center',color='blue')
  # i e j: Posição do texto em relação as barras (aproximam ou afastam o texto das barras)
  i = 1.0
  j = 100
  # impressão das legendas
  for i in range(len(tipoImovel)):
      plt.annotate(contTipoImovel[i], (-0.1 + i, contTipoImovel[i] + j))
  # legenda do gráfico
  plt.legend(labels = [legendaGrafico])
  # Título
  plt.title(tituloGrafico)
  # Legendas eixo X e Y
  plt.xlabel(rotuloEixoX)
  plt.ylabel(rotuloEixoY)
  # Exibir o gráfico
  plt.show()
############## FIM DA FUNÇÃO ##################

# chamada da função com os parâmetros para montar o gráfico
graficoBarra('room_type','Total por tipo de imóvel','Tipo de imóvel','Total','Tipo de Imóvel')
```
